## Finding PKs

```python
total_rows = len(df_trials_unique)

for col in df_trials_unique.columns:
    unique = df_trials_unique[col].nunique(dropna=False)
    print(f"{col:30} {unique:,} / {total_rows:,}")
```

```python
def test_key(df, cols):
    duplicates = (
        df
        .groupby(cols, dropna=False)
        .size()
        .gt(1)
        .sum()
    )

    print(f"{cols}")
    print(f"Duplicate groups: {duplicates}")
    print(f"Candidate key: {'YES' if duplicates == 0 else 'NO'}")
```

```python
test_key(df_trials_unique, ["Brief Title", "Organization Full Name"])
```

```python
dups = (
    df_trials_unique
    .groupby(
        [
            "Brief Title",
            "Organization Full Name",
            "Full Title",
            "Start Date",
            "Interventions",
            "Conditions",
            "Phases",
            "Outcome Measure",
            "Intervention Description"
        ],
        dropna=False,
    )
    .size()
    .reset_index(name="count")
    .query("count > 1")
)

dups
```

```python
cols = [
    "Brief Title",
    "Organization Full Name",
    "Full Title",
    "Start Date",
    "Interventions",
    "Conditions",
    "Phases",
    "Outcome Measure",
    "Intervention Description"
]

duplicate_rows = df_trials_unique.merge(
    dups[cols],
    on=cols,
    how="inner"
)

duplicate_rows
```
